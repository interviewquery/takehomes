## Problem Statement


Two years ago, MAPMG opened a new inpatient wing at the Maryland Hospital for Data Scientists (MHDS).
For the first few months, everything was fine - but then, some of the new physicians began reporting
that they were having days where they were getting overwhelmed.  The hospital is currently staffed
with 15 doctors, each of whom is expected to work with about 10 patients per day.  Since the average
patient census (number of patients in the hospital on a given day) is around 150, some members of the management
team are inclined to disregard these complaints.

Physicians and management have asked you to investigate whether there are strong indicators that the
physicians are getting overworked, and they've settled on a metric that is important to the bottom line:
patient discharges.  Some physicians have reported that the discharge process can be very difficult
to manage during busy days.  This can lead to a spiral where physicians have to triage more and more
cases, preventing them from discharging complex patients, but then they have even more patients to deal
with the next day.

The lead physician at the hospital would like you to investigate whether the amount
of work at the hospital (i.e. the number of patients) affects the discharge rate.  If so, then 
management would like to know how large an effect this has, since that will determine how cost
effective it would be to call in additional physicians on busy days.

The physicians have given you the following directives to help guide your analysis:
- Some patients are relatively simple, and these cases take very little time to discharge;
you can expect that patients with only a few days in the hospital won't be affected.
- The physicians believe that they're really having trouble when they have to deal with 10
or more patients each, and 8 patients each feels like a very manageable number.

You have a dataset (`hospital_sim`) with patient admission and discharge dates (encoded as integers which count the days
from the hospital's opening).  This data includes records for 20,000 patients over the two year period.
Unfortunately, the data that you have available right now doesn't include information about
the patient's health state or why the patient was in the hospital.
